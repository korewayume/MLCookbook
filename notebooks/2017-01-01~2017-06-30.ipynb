```python
RASTER_FILE_PATH = 'images/2017'
RASTER_REGEX = re.compile(r'^2017-(?P<month>\d{2})-(?P<day>\d{2}).tif$')
def ReadBandArray(raster_filename, band_number=1):
    try:
        raster = gdal.Open(raster_filename)
        band = raster.GetRasterBand(band_number)
        return band.ReadAsArray()
    except Exception as e:
        print e
        print raster_filename, os.path.exist(raster_filename)

raster_filenames = []
for dirpath, dirnames, filenames in os.walk(RASTER_FILE_PATH):
    for filename in filenames:
        if re.match(RASTER_REGEX, filename):
            raster_filenames.append(os.path.join(dirpath, filename))
```

# 读取数据

```python
fit_data = None
for filename in raster_filenames:
    a = ReadBandArray(filename).reshape(-1,1)
    if fit_data is None:
        fit_data = a
    else:
        fit_data = np.hstack((fit_data, a))


fit_data[np.isnan(fit_data)] = -1

fit_data.shape
```

# k-Means聚类 n_clusters=8

```python
from sklearn import  cluster
n_clusters=8
clst=cluster.KMeans(n_clusters=n_clusters)
clst.fit(fit_data)
```

# 曲线拟合示例

```python
from scipy.optimize import leastsq
one_curve = fit_data[:,np.equal(result, 4)][:,1]
xArr = np.linspace(0,one_curve.shape[0],one_curve.shape[0])
polynomial_degree = 10


def func(x, p):
    """
    数据拟合所用的函数
    """

    f = np.poly1d(p)
    return f(x)


def residuals(p, y, x, n):
    """
    实验数据x, y和拟合函数之间的差，p为拟合需要找到的系数
    """
    
    ret = y - func(x, p)
    return ret


p0 = np.array([1] * polynomial_degree)
plsq = leastsq(residuals, p0, args=(one_curve, xArr, polynomial_degree))
plt.subplot(211)
plt.plot(xArr, one_curve, '*', label=u'原始数据')
plt.plot(xArr, func(xArr, plsq[0]), "r", label=u"拟合曲线")
plt.legend()
plt.subplot(212)
plt.plot(xArr, func(xArr, np.polyder(plsq[0])), "r", label=u"一阶导数曲线")
plt.legend()
plt.show()
```

```python
for i in range(clst.cluster_centers_.shape[0]):
    curve = clst.cluster_centers_[i]
    x = np.linspace(0, curve.shape[0], curve.shape[0])
    plsq = leastsq(residuals, p0, args=(curve, x, polynomial_degree))
    plt.subplot(211)
    plt.plot(x, curve, '*', color=colors[i], label=u'原始数据')
    plt.plot(x, func(x, plsq[0]), "r", label=u"拟合曲线")
    plt.grid(True)
    plt.legend()
    plt.subplot(212)
    plt.plot(x, func(x, np.polyder(plsq[0])), "r", label=u"一阶导数曲线")
    plt.grid(True)
    plt.legend()
    plt.show()
    
```
