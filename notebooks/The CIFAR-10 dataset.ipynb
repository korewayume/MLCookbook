```python
# -*- coding: utf-8 -*-
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
np.set_printoptions(precision=4, threshold=500)
pd.options.display.max_rows = 100
%matplotlib inline
```

```python
label_names = [
    "airplane",
    "automobile",
    "bird",
    "cat",
    "deer",
    "dog",
    "frog",
    "horse",
    "ship",
    "truck"
]

ch_label_names = [
    "飞机",
    "汽车",
    "鸟",
    "猫",
    "鹿",
    "狗",
    "青蛙",
    "马",
    "船",
    "卡车"
]
def load_dataset(batch_no):
    import cPickle
    with open("cifar-10-batches-py/data_batch_" + str(batch_no), 'rb') as fo:
        cifar_10 = cPickle.load(fo)
    return cifar_10

def show_image_data(image):
    return np.rot90(image.reshape((32,32,3), order="F"), -1)
```
